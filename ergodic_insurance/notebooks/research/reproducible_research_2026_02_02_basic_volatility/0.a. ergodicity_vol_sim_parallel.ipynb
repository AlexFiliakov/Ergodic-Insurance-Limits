```python
from tqdm.notebook import tqdm  # progress bar
from itertools import product
import os
from joblib import Parallel, delayed
import shutil # Import shutil for file operations

# Integrate joblib with tqdm for per-task progress
class TqdmJoblib(tqdm):
    def __enter__(self):
        from joblib import parallel
        self._original = parallel.BatchCompletionCallBack
        outer = self
        class TqdmBatchCompletionCallback(parallel.BatchCompletionCallBack):
            def __call__(self, *args, **kwargs):
                outer.update(n=self.batch_size)
                return super().__call__(*args, **kwargs)
        parallel.BatchCompletionCallBack = TqdmBatchCompletionCallback
        return super().__enter__()
    def __exit__(self, exc_type, exc, tb):
        from joblib import parallel
        parallel.BatchCompletionCallBack = self._original
        return super().__exit__(exc_type, exc, tb)

NUM_SIMULATIONS = 250_000
SIM_YEARS = 50
PRICING_SIMULATIONS = 1_000_000  # For premium estimation
CRN_BASE_SEED = 20260130  # Common Random Numbers seed (date: 2026-01-30)

# Test override
NUM_SIMULATIONS = 1_000
SIM_YEARS = 5
PRICING_SIMULATIONS = 50_000  # For premium estimation

initial_assets_list = [5_000_000, 7_500_000, 10_000_000, 15_000_000, 25_000_000]
atr_list = [0.8, 0.9, 1.0, 1.1, 1.2]
ebitabl_list = [0.1, 0.125, 0.15]
deductible_list = [0, 50_000, 100_000, 250_000, 350_000, 500_000, 750_000]
loss_ratio_list = [0.6, 0.7, 0.8]
volatility_list = [0.15, 0.2, 0.25]

# Reduce simulations to run in time
initial_assets_list = [5_000_000, 10_000_000, 25_000_000]
atr_list = [0.8, 1.0, 1.2]
ebitabl_list = [0.125]
deductible_list = [0, 100_000, 250_000, 500_000]
loss_ratio_list = [0.7]
volatility_list = [0.15]


total_combos = (
    len(initial_assets_list)
    * len(atr_list)
    * len(ebitabl_list)
    * len(deductible_list)
    * len(loss_ratio_list)
    * len(volatility_list)
)

# Use all but one CPU core
_cpu = os.cpu_count() or 1
n_jobs = max(_cpu, 1)

print(f'Available cores: {n_jobs}')
print(f'Total parameter combinations: {total_combos}')
print(f'Simulations per config: {NUM_SIMULATIONS:,}')
print(f'Simulation horizon: {SIM_YEARS} years')
print(f'CRN base seed: {CRN_BASE_SEED}')

# Define the destination directory in your Google Drive
# Make sure 'My Drive/results' exists in your Google Drive
gdrive_results_dir = '/content/drive/My Drive/Colab Notebooks/results_vol_full'
os.makedirs(gdrive_results_dir, exist_ok=True) # Create the directory if it doesn't exist

# Parallel execution with progress bar
with TqdmJoblib(total=total_combos, desc="Parameter sets", leave=True):
    Parallel(n_jobs=n_jobs, prefer="processes", batch_size=1)(
        delayed(run_vol_sim)(
            index=index,
            ia=ia,
            atr=atr,
            ebitabl=ebitabl,
            ded=ded,
            lr=lr,
            NUM_SIMULATIONS=NUM_SIMULATIONS,
            SIM_YEARS=SIM_YEARS,
            PRICING_SIMULATIONS=PRICING_SIMULATIONS,
            filepath=gdrive_results_dir,
            volatility=vol,
            crn_base_seed=CRN_BASE_SEED,
        )
        for index, (ia, atr, ebitabl, ded, lr, vol) in enumerate(product(
            initial_assets_list,
            atr_list,
            ebitabl_list,
            deductible_list,
            loss_ratio_list,
            volatility_list
        ))
    )
```
